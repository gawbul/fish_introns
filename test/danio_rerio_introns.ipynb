```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

danio_rerio_introns_df = pd.read_csv("danio_rerio_introns.csv")
bins = list(range(0, danio_rerio_introns_df["Length"].max(), 30))
introns_data = (
    danio_rerio_introns_df["Length"]
    .value_counts(bins=bins, sort=False)
    .rename_axis("Intron Length")
    .reset_index(name="Frequency")
)
introns_data["Intron Length"] = introns_data.apply(
    lambda row: (row["Intron Length"].left + row["Intron Length"].right) / 2, axis=1
)
plot = sns.relplot(
    data=introns_data,
    x="Intron Length",
    y="Frequency",
    kind="line",
    linewidth=0.5,
)
plt.xlim(0, 5000)
plot.set(title="Frequency of intron lengths in Danio rerio")
plot.set(xlabel="Intron Length")
plot.set(ylabel="Frequency")
plot.set(yscale="log")
plot.figure.savefig("danio_rerio_introns.png", dpi=600)
```
